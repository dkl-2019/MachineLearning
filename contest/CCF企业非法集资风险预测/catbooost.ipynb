```python
import pandas as pd
import numpy as np
```

```python
base = pd.read_csv(open("train/base_info.csv",encoding='utf-8'))
train_id = pd.read_csv(open("train/entprise_info.csv",encoding='utf-8'))
test_id = pd.read_csv(open("entprise_evaluate.csv",encoding='utf-8'))
```

```python
concat_id = train_id.append(test_id)
concat_id = concat_id.drop("score",axis=1)
```

```python
data1 = pd.merge(base,concat_id,how='outer',on='id')
```

```python
train = data1[data1['label'].notnull()]
test = data1[data1['label'].isnull()]
```

```python
train.info()
```

```python
test.info()
```

```python
train.head()
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
train['industryphy'] = le.fit_transform(train['industryphy'])
train = train.drop(['ptbusscope','midpreindcode'],axis=1)
test['industryphy'] = le.fit_transform(test['industryphy'])
test = test.drop(['ptbusscope','midpreindcode'],axis=1)
```

```python
train.info()
```

```python
train_data = train.drop(['label','dom','id','opscope','opfrom','opto','opform','oploc'],axis=1)
y_train = train['label']
```

```python
test_data = test.drop(['label','dom','id','opscope','opfrom','opto','opform','oploc'],axis=1)
```

```python
def precess_func(train):
    train['enttypeitem'] = train['enttypeitem'].fillna(train['enttypeitem'].mode())
    train = train.drop(['protype','forreccap','forregcap','congro'],axis=1)
    return train
train_data = precess_func(train_data)
test_data = precess_func(test_data) 
```

```python
train_data.info()
```

```python
from sklearn import metrics
import xgboost as xgb
from sklearn.model_selection import train_test_split
X_train_split,X_val,y_train_split,y_val = train_test_split(train_data,y_train,test_size=0.2,random_state=2020)
train_martix = xgb.DMatrix(X_train_split,label=y_train_split)
val_martix = xgb.DMatrix(X_val,label=y_val)
params = {
        'booster': 'gbtree',
        'objective': 'binary:logistic',
        'eval_metric': 'auc',
        'gamma': 1,
        'min_child_weight': 1.5,
        'max_depth': 5,
        'lambda': 10,
        'subsample': 0.7,
        'colsample_bytree': 0.7,
        'colsample_bylevel': 0.7,
        'eta': 0.04,
        'tree_method': 'exact',
        'seed': 2020,
        'n_jobs': -1,
        'silent': 0,
    }
# watchlist = [(train_martix,"train"),(val_martix,"val")]
model = xgb.train(params,train_martix,num_boost_round=20000,verbose_eval=500)
val_pred = model.predict(val_martix,ntree_limit=model.best_ntree_limit)
```

```python
from sklearn.metrics import roc_auc_score
roc_auc_score(y_val,val_pred)
```

```python
val_pred1 = []
for i in val_pred:
    if i>0.5:
        a = 1
        val_pred1.append(a)
    else:
        a = 0
        val_pred1.append(a)
```

```python
val_pred.max()
```

```python
from sklearn.metrics import f1_score
f1_score(y_val,val_pred1)
```

```python
test_Dmarix = xgb.DMatrix(test_data)
xgb_pre = model.predict(test_Dmarix)
```

```python
submission = pd.read_csv('entprise_submit.csv')
submission.score = xgb_pre
submission.to_csv('1101_2.csv',index=False)
```
