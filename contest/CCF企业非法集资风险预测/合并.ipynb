```python
from sklearn.preprocessing import LabelEncoder
le1  = LabelEncoder()
le2 = LabelEncoder()
tax_info['TAX_CATEGORIES'] = le1.fit_transform(tax_info['TAX_CATEGORIES'])
tax_info['TAX_ITEMS'] = le2.fit_transform(tax_info['TAX_ITEMS'])
```

```python
tax_info['START_DATE'] = pd.to_datetime(tax_info['START_DATE'],format="%Y-%m-%d")
tax_info['END_DATE'] = pd.to_datetime(tax_info['END_DATE'],format="%Y-%m-%d")
```

```python
tax_info['Time'] = tax_info['END_DATE'] - tax_info['START_DATE']
```

```python
tax_info['Time'] = tax_info['Time'].apply(lambda x:x.days)
```

```python
id_list2 = tax_info.id.unique()
tax_categories_list = []
tax_items_list = []
taxation_basis_list = []
tax_rate_list = []
tax_deduction_list = []
tax_amount_list = []
tax_time_list = []
for i in id_list2:
    tax_categories = tax_info[tax_info['id']==i]['TAX_CATEGORIES'].values
    tax_categories_list.append(tax_categories)
    tax_items = tax_info[tax_info['id']==i]['TAX_ITEMS'].values
    tax_items_list.append(tax_items)
    taxation_basis= tax_info[tax_info['id']==i]['TAXATION_BASIS'].values
    taxation_basis_list.append(taxation_basis)
    tax_rate = tax_info[tax_info['id']==i]['TAX_RATE'].values
    tax_rate_list.append(tax_rate)
    tax_deduction = tax_info[tax_info['id']==i]['DEDUCTION'].values
    tax_deduction_list.append(tax_deduction)
    tax_amount = tax_info[tax_info['id']==i]['TAX_AMOUNT'].values
    tax_amount_list.append(tax_deduction)
    tax_time = tax_info[tax_info['id']==i]['Time'].values
    tax_time_list.append(tax_time)
```

```python
TAX_CATEGORIES = tax_info.groupby("id").count()['TAX_CATEGORIES'].values
TAXATION_BASIS = tax_info.groupby("id").mean()['TAXATION_BASIS'].values
TAX_RATE = tax_info.groupby("id").mean()['TAX_RATE'].values
DEDUCTION = tax_info.groupby("id").mean()['DEDUCTION'].values
TAX_AMOUNT = tax_info.groupby("id").mean()['TAX_AMOUNT'].values
```

```python
tax_data = {
    "id":tax_info.id.unique(),
    "TAX_CATEGORIES":tax_categories_list,
    'TAX_ITEMS':tax_items_list,
    'TAXATION_BASIS':taxation_basis_list,
    "TAX_RATE":tax_rate_list,
    "DEDUCTION":tax_deduction_list,
    "TAX_AMOUNT":tax_amount_list,
    'Time':tax_time_list
}
tax_info1 = pd.DataFrame(tax_data)
```

```python
data = pd.merge(data,tax_info1,on="id",how='outer')
```
